```python
import pandas as pd
from pandasql import sqldf
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
sqldf("select * from df limit 10")
```

```python
sqldf("select location,new_cases from df limit 10")
```

```python
sqldf("SELECT location,new_cases FROM df WHERE  location = 'Uruguay' LIMIT 10")
```

```python
sqldf("SELECT location,new_cases,total_deaths, date  FROM df WHERE  location = 'Uruguay' and date >'2021-01-01' ORDER BY date ASC LIMIT 10")
```
